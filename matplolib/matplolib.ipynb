```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
names = list(iris.target_names)
print(f'x contient {x.shape[0]} examples et {x.shape[1]} variables')
print(f'il y a {np.unique(y).size} classe ')
y
```

# Graphique de classification

```python
plt.scatter(x[:,0], x[:,1])
```

```python
#representation des données selon la nature de leur classe
plt.scatter(x[:,0], x[:,1] , c=y) #c pour la couleur
#Donc on aura 3 couleur
plt.xlabel('longeur sépal')
plt.ylabel('largeur sépal')
```

### Astuse sur le scatter plot

#### Astuce 1 : le paramètre alpha : permet de controler la transparence de chacun des points de notre graphe 
#### Astuce 2 : le parametre s qui permet de controller la taille de nos points en fonction des variables de notre problèmes

```python
plt.scatter(x[:,0], x[:,1] , c=y , alpha=0.5 , s = 100) 

plt.xlabel('longeur sépal')
plt.ylabel('largeur sépal')
```
